```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
#Loading all data into individual data frames
customers = pd.read_excel("bikeshops.xlsx",sheetIndex=1)
orders = pd.read_excel("orders.xlsx",sheetIndex=1)
products = pd.read_excel("bikes.xlsx",sheetIndex=1)

customers.head()
# orders.head()
# products.head()
```

```python
#Combining above data into a single dataframe

ordersXcustomers = pd.merge(orders,customers,left_on = "customer.id",right_on = "bikeshop.id")
combinedDF = pd.merge(ordersXcustomers,products,left_on = "product.id", right_on = "bike.id" )

combinedDF.head()
```

```python
#Calculate total value of each order
combinedDF['price.extended'] = combinedDF['price'] * combinedDF['quantity']
#Sort the resultant by order.id and order.line
combinedDF.sort_values(['order.id','order.line'],inplace=True)


#Drop columns unnecessary for clustering
trimmedDF = combinedDF.drop(['customer.id','product.id','bikeshop.id','bikeshop.city','bikeshop.state','latitude','longitude','bike.id','index'],axis = 1)

trimmedDF.head()
```

```python
customerTrends = trimmedDF.groupby(['bikeshop.name', 'model', 'category1', 'category2', 'frame', 'price']).agg({'quantity':'sum'})

customerTrends = customerTrends.pivot_table(index = ['model','category1','category2','frame','price'],columns = 'bikeshop.name',values = 'quantity')
customerTrends.fillna(0,inplace=True)

customerTrends.head()
```

```python
#Normalizing columns
for column in customerTrends.columns:
    customerTrends[column] =  (customerTrends[column]/customerTrends[column].sum())
```

```python
#Converting indicies to columns
customerTrends.reset_index(level = customerTrends.index.names,inplace = True)
customerTrends.head()
```

```python
#Categorizing prices column into 2 bins
x = pd.cut(customerTrends['price'],bins = 2,retbins=True)
customerTrends['price'] = x[0]
#print(customerTrends[customerTrends['model']=='Scalpel-Si Carbon 3'])
customer_perms = customerTrends.iloc[:,:5]
```

```python
#K-means
kmeansDat = customerTrends.iloc[:,5:]
customers = kmeansDat.columns.values

kmeansDat = kmeansDat.transpose()
#print(kmeansDat.head())
number_of_clusters = 5
kmeans = KMeans(n_clusters=number_of_clusters)
model = kmeans.fit(X= kmeansDat)
labels = model.labels_
```

```python
#Analyze clustering results
clusters = {}

for index,label in enumerate(labels):
    if label in clusters:
        clusters[label].append(customers[index])
    else:
        clusters[label] = [customers[index]]

sorted_clusters = sorted(clusters.items(),key = lambda x : x[0])
```

```python
print("Following clusters were found\n")
for tup in sorted_clusters:
    print("Cluster :" + str(tup[0]))
    print("--------------------")
    for customer in tup[1]:
        print(customer)
    print("--------------------\n")
```

```python
#Map clusters to the bike models
centers = model.cluster_centers_.transpose()
```

```python
cluster_number =[]
cluster_dist = []
for c in centers:
    bestc = max(c)  #gets the best cluster centroid
    cluster_number.append(list(c).index(bestc))
    cluster_dist.append(bestc)
customer_perms['cluster'] = cluster_number
customer_perms['center'] = cluster_dist
```

```python
for i in range(0,number_of_clusters):
    cluster1 = customer_perms[customer_perms['cluster']==i].sort_values(['center'],ascending=[False])
    cluster1.reset_index(inplace=True)
    print (cluster1.iloc[0:10,:])
```
